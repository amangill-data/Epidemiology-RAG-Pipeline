```python
from google.cloud import storage

client = storage.Client()

for blob in client.list_blobs(BUCKET_NAME, prefix=folder_prefix):
  print(str(blob))
```

```python
import os
from langchain_community.document_loaders import PyPDFLoader

print(f"Processing documents from gs://{BUCKET_URI}")

bucket = storage_client.bucket(BUCKET_NAME)

# Create a temporary local folder in Colab to hold the PDFs
os.makedirs("./temp_pdfs", exist_ok=True)

# load documents
all_documents = []
blobs = bucket.list_blobs(prefix=folder_prefix)

for blob in blobs:
    print(str(blob))

    if blob.name.endswith("/") or not blob.name.lower().endswith(".pdf"):
        continue

    print(f" Loading document: {blob.name}")

    # 1. Download the PDF locally first (this takes milliseconds)
    document_name = blob.name.split("/")[-1]
    local_file_path = f"./temp_pdfs/{document_name}"
    blob.download_to_filename(local_file_path)

    # 2. Use PyPDFLoader instead of GCSFileLoader (bypasses the heavy OCR)
    loader = PyPDFLoader(local_file_path)
    documents_from_blob = loader.load()

    doc_source_prefix = f"gs://{BUCKET_NAME}"
    doc_source_suffix = "/".join(blob.name.split("/")[0:-1])
    source = f"{doc_source_prefix}/{doc_source_suffix}"

    for document in documents_from_blob:
        document.metadata["source"] = source
        document.metadata["document_name"] = document_name

        all_documents.append(document)

print(f"# of document pages loaded (pre-chunking) = {len(all_documents)}")
```

```python
import time
from google.api_core.exceptions import ResourceExhausted

# (Make sure you've run the cell that creates doc_splits before this!)
texts = [doc.page_content for doc in doc_splits]
metadatas = [doc.metadata for doc in doc_splits]

# Lowered batch size to avoid hitting the Tokens Per Minute (TPM) limit
batch_size = 25
max_retries = 5

for i in range(0, len(texts), batch_size):
    batch_texts = texts[i : i + batch_size]
    batch_metadatas = metadatas[i : i + batch_size]
    print(f"Adding batch {i//batch_size + 1}/{(len(texts)+batch_size-1)//batch_size} (size: {len(batch_texts)})...")

    retries = 0
    while retries < max_retries:
        try:
            # This step embeds the text AND uploads to Vector Search
            vsvectordb.add_texts(texts=batch_texts, metadatas=batch_metadatas)
            break # Break out of the while loop if successful!

        except ResourceExhausted as e:
            # This catches the specific Quota/Rate Limit error
            retries += 1
            # Exponential backoff: sleeps for 10s, then 20s, then 40s...
            sleep_time = (2 ** retries) * 5
            print(f"   [!] Quota hit! Sleeping for {sleep_time} seconds before retrying (Attempt {retries}/{max_retries})...")
            time.sleep(sleep_time)

        except Exception as e:
            # Catch other random errors so your whole loop doesn't crash
            print(f"   [!] An unexpected error occurred: {e}")
            break

    # A tiny base delay between successful batches to pace out the API requests
    time.sleep(2)

print("All texts processed.")
```

```python
def formatter(result):
  print(f"Query: {result['input']}")
  print("."*80)
  if "context" in result.keys():
    for idx, ref in enumerate(result["context"]):
      print("-"*80)
      print(f"REFERENCE #{idx}")
      print("-"*80)
      if "score" in ref.metadata:
        print(f"Matching Score: {ref.metadata['score']}")
      if "source" in ref.metadata:
        print(f"Document Source: {ref.metadata['source']}")
      if "document_name" in ref.metadata:
        print(f"Document Name: {ref.metadata['document_name']}")
      print("."*80)
      print(f"Content: \n{wrap(ref.page_content)}")
  print("."*80)
  print(f"Response: {wrap(result['answer'])}")
  print("."*80)
```

```python
def wrap(s):
  return "\n".join(textwrap.wrap(s, width=120, break_long_words=False))
```

```python
def ask(
    query,
    k=NUMBER_OF_RESULTS,
    # search_distance_threshold=SEARCH_DISTANCE_THRESHOLD,
    filters=None,
):
  retriever.search_kwargs["k"] = k

  if filters:
    ns_filters = [Namespace(name=filters["namespace"],
                            allow_tokens=filters["allow_list"])]
    retriever.search_kwargs["filter"] = ns_filters

  else:
    retriever.search_kwargs.pop("filter", None)


  result = retrieval_qa_chain.invoke({"input": query})
  return formatter(result)
```
